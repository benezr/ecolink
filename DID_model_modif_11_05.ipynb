### Resource efficiency actions (q1_, q2_)

```python
# Resource efficiency actions (several features) based on q1 and q2, dummies, NANs = 0
# New features: 0-not planning, 1-planning, 2-already in action
def to_binary_action(series: pd.Series) -> pd.Series:
    # numeric case
    if pd.api.types.is_numeric_dtype(series):
        return (series.fillna(0) != 0).astype('int8')
    # string case
    s = series.astype('string').str.strip()
    return ((s.notna()) & (s != "Not mentioned")).astype('int8')
```

```python
domain_pairs = {
    'water': ('q1_1','q2_1'),
    'energy_saving': ('q1_2','q2_2'),
    'green_energy': ('q1_3','q2_3'),
    'materials': ('q1_4','q2_4'),
    'green_suppliers': ('q1_5','q2_5'),
    'waste_reduction': ('q1_6','q2_6'),
    'sell_residues': ('q1_7','q2_7'),
    'recycling': ('q1_8','q2_8'),
    'eco_design': ('q1_9','q2_9'),
    'other': ('q1_10','q2_10')
}
```

```python
# convert only the involved columns
cols_to_convert = sorted({c for pair in domain_pairs.values() for c in pair})
for c in cols_to_convert:
    df[c] = to_binary_action(df[c])
```

```python
for name, (cur, plan) in domain_pairs.items():
    df[f'{name}_maturity'] = np.select(
        [
            (df[cur] == 0) & (df[plan] == 0),   # none
            (df[cur] == 0) & (df[plan] == 1),   # planned only
            (df[cur] == 1)                      # implemented (whether planned or not)
        ],
        [0, 1, 2],
        default=0
    ).astype('int8')
```

```python
for name, (cur, plan) in domain_pairs.items():
    df[f'{name}_maturity_d'] = np.select(
        [
            (df[cur] == 0) & (df[plan] == 0),   # none
            (df[cur] == 0) & (df[plan] == 1),   # planned only
            (df[cur] == 1)                      # implemented (whether planned or not)
        ],
        [0, 1, 1],
        default=0
    ).astype('int8')
```
